```python
import pandas as pd
```

```python
def points_to_df(array_of_points):
  """ Function to to take array of points generated by the plot digitizer to a data frame with columns frequency and amplitude"""
  tuples = [(array_of_points[i], array_of_points[i+1]) for i in range(0, len(array_of_points),2)]
  return pd.DataFrame(tuples, columns=['Frequency', 'Amplitude'])
```
